```python
import numpy as np
```

```python
def gaussSeidel(A, x ,b):    
    n = len(A)                    

    for j in range(0, n):         
        d = b[j]   
        
        for i in range(0, n):      
            if(j != i): 
                d -= A[j][i] * x[i] 

        x[j] = d / A[j][j] 
        
    return x     
```

```python
A = np.zeros((128, 128))
e = np.ones(128)
x = np.zeros(128)
for i in range(128):
    A[i, i] = 4

for i in range(127):
    A[i + 1, i] = 1
    A[i, i + 1] = 1

for i in range(124):
    A[i + 4, i] = 1
    A[i, i + 4] = 1

for i in range(100):
    print(x)
    x = gaussSeidel(A, x, e)
    
```
