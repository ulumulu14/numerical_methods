```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import interpolate
```

```python
data = pd.read_csv("zad20data.txt", sep = '\t', lineterminator = '\n', names = ['x', 'y'])
x = np.array(data.x)
y = np.array(data.y)
```

```python
tck = interpolate.splrep(x, y, s = 0)
y_new = interpolate.splev(x, tck, der = 0)
```

```python
plt.figure(figsize = (14, 10))
plt.plot(x, y, '.', x, y_new)
```
