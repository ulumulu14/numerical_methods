```python
import numpy as np
from scipy.sparse.linalg import cg
```

```python
def conjugateGradient(A, x, b):
    n = len(b)
    
    r = np.dot(A, x) - b
    p = - r
    r_k_norm = np.dot(r, r)
    
    for i in range(2 * n):
        Ap = np.dot(A, p)
        alpha = r_k_norm / np.dot(p, Ap)
        x += alpha * p
        r += alpha * Ap
        r_kplus1_norm = np.dot(r, r)
        beta = r_kplus1_norm / r_k_norm
        r_k_norm = r_kplus1_norm
        
        if r_kplus1_norm < 1e-5:
            break
            
        p = beta * p - r
        
    return x
```

```python
A = np.zeros((128, 128))
e = np.ones(128)
x = np.zeros(128)
for i in range(128):
    A[i, i] = 4

for i in range(127):
    A[i + 1, i] = 1
    A[i, i + 1] = 1

for i in range(124):
    A[i + 4, i] = 1
    A[i, i + 4] = 1

x = conjugateGradient(A, x, e)
print(x)
```
