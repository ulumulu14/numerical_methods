```python
import numpy.linalg
from scipy.optimize import minimize
```

```python
lambd = (1.0/1024)
```

```python
def hessian(x):
    global lambd
    return [[(1+lambd)*(2-2*x[0]+100*(-4*x[1]*x[0]+4*pow(x[0],3))), -400*x[0]], [-400*x[0], (1+lambd)*(200)]]
```

```python
def gradient(x):
    return [2*(1-x[0])-400*x[0]*(-x[0]*x[0]+x[1]), 200*(-x[0]*x[0]+x[1])]
```

```python
def rosenbrock(p):
    return((1-p[0])*(1-p[0]) + 100*(p[1]-p[0]*p[0])*(p[1]-p[0]*p[0]))
```

```python
def rosenbrockDer(x):
    xm = x[1:-1]
    xm_m1 = x[:-2]
    xm_p1 = x[2:]
    der = numpy.zeros_like(x)
    der[1:-1] = 200*(xm-xm_m1**2) - 400*(xm_p1 - xm**2)*xm - 2*(1-xm)
    der[0] = -400*x[0]*(x[1]-x[0]**2) - 2*(1-x[0])
    der[-1] = 200*(x[-1]-x[-2]**2)
    return der
```

```python
def printer(x):
    print(x)
```

```python
def lev_mar(x):
    global lambd
    grad_flag = True
    hess_flag = True
    grad = [0.0, 0.0]
    hess = [[0.0, 0.0], [0.0, 0.0]]
    x_test = [1.0, 1.0]
    prev_x = [1.0, 1.0]
    i = 0
    
    while i < 10:
        if grad_flag == True:
            grad = gradient(x)

        if hess_flag == True:
            hess = numpy.linalg.inv(hessian(x))
            x_test[0] = x[0]-(hess[0][0]*grad[0]+hess[0][1]*grad[1])
            x_test[1] = x[1]-(hess[1][0]*grad[0]+hess[1][1]*grad[1])
            
        if lambd > 1.0:
            print("Poza basenem abstrakcji minimum")
            break
            
        if lambd == 0.0:
            print("Macierz H prawie rowna hessjanowi")
            print(minimize(rosenbrock, x, callback=printer,method='nelder-mead').x)
            break
            
        if rosenbrock(x_test) > rosenbrock(x):
            lambd = lambd*8
            grad_flag = False
            #i=i+1
            continue
            
        if rosenbrock(x_test) < rosenbrock(x):
                lambd = lambd/8
                print(x)
                
        if abs(abs(x[0])-abs(prev_x[0])) < 1e-7:
            break
            
        print("Podmiana")
        
        prev_x[0] = x[0]
        prev_x[1] = x[1]
        x[0] = x_test[0]
        x[1] = x_test[1]

        print(x)
        print(lambd)
        
        grad_flag = True
        hess_flag = True

        continue
```
